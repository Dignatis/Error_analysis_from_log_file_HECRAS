```python
import pandas as pd
import numpy as np
import os 
```

## 1.0 Opening log file

```python
logfile = 'Log_file_1.txt'
with open(logfile) as f:
    all_lines = f.readlines()
```

## 1.1 Selecting only error lines

```python
with open ('errors.txt' , 'w') as errors:
    n = 0 
    for l in all_lines:
        if 'Cell #' in l:
            n = n + 1
            errors.write(l)
```

## 1.2 Errors txt to DF

```python
df = pd.read_csv('errors.txt', sep='\t', lineterminator='\n', header=None)
df = df.rename({0:'Maximum iteration location', 1:'Cell#',
               2:'Cell_Numb', 3:'WSEL', 4:'Error', 5:'Iterations'}, axis='columns')
df = df.drop('Cell#', axis=1)
df
```

## 1.3 Deleting temporary file

```python
os.remove('errors.txt')
```

## 2.1 List of cells with error

```python
err_cells = df['Cell_Numb'].unique()
```

## 2.1 Empty list for stats of cell

```python
sum_err = []
avg_err = []
cou_err = []
```

## 2.3 Stats for each cell 


Napraviti da prikazuje top 10 najvećih grešaka po SUM
<br>
Da filtrira greške po nekom kriterijumu


```python
for i in err_cells:
    df_ec = df.loc[df['Cell_Numb'] == int(i)]
    
    sum_err_cell = df_ec['Error'].sum()
    sum_err.append(sum_err_cell)
    
    avg_err_cell = df_ec['Error'].mean()
    avg_err.append(avg_err_cell)
    
    cou_err_cell = df_ec['Error'].count()
    cou_err.append(cou_err_cell)   
```

```python
df_cell_stats = pd.DataFrame({'Cell':err_cells, 'SUM error':sum_err, 'Average error':avg_err, 'Number of errors':cou_err})
df_cell_stats
```

## 3.1 Analysis - cells with bigest SUM error

```python
df_cell_stats.nlargest(10, ['SUM error'])
```

## 3.2 Analysis - cells with bigest average error

```python
df_cell_stats.nlargest(10, ['Average error'])
```

## 3.3 Analysis - cells with highest occurrence of error

```python
df_cell_stats.nlargest(10, ['Number of errors'])
```
